```python
points_local=[]

for i in df_new.index:
  points_local.append(0)


for i in range(1,1521):
   games=0
   for x in range(-len(df_new.index),-i):
     if(df_new.HOME_TEAM[i]==df_new.HOME_TEAM[-x] and df_new.AWAY_TEAM[i]==df_new.AWAY_TEAM[-x]):
        games=games+1
        if(df_new.FTR[-x]=='H'and games==1):
          points_local[i-1]=3
        if(df_new.FTR[-x]=='D' and games==1 ):
           points_local[i-1]=1
        else:
          if(df_new.FTR[-x]=='H'):
             points_local[i-1]=3*0.3 +0.7*points_local[i-1]
          if(df_new.FTR[-x]=='D'):
             points_local[i-1]=1*0.3 +0.7*points_local[i-1]
          if(df_new.FTR[-x]=='A'):
             points_local[i-1]=0.7*points_local[i-1]
  
   if(games<3):
     points_local[i-1]=-33
```

```python
away_points=[]

for i in df_new.index:
  away_points.append(0)


for i in range(1,1521):
   games=0
   for x in range(-len(df_new.index),-i):
     if(df_new.HOME_TEAM[i]==df_new.HOME_TEAM[-x] and df_new.AWAY_TEAM[i]==df_new.AWAY_TEAM[-x]):
        games=games+1
        if(df_new.FTR[-x]=='A'and games==1):
          away_points[i-1]=3
        if(df_new.FTR[-x]=='D' and games==1 ):
           away_points[i-1]=1
        else:
          if(df_new.FTR[-x]=='A'):
             away_points[i-1]=3*0.3 +0.7*away_points[i-1]
          if(df_new.FTR[-x]=='D'):
             away_points[i-1]=1*0.3 +0.7*away_points[i-1]
          if(df_new.FTR[-x]=='H'):
             away_points[i-1]=0.7*away_points[i-1]
  
   if(games<3):
     away_points[i-1]=-33
```
