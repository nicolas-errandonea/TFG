```python
import statistics
```

```python
def home_mean(i,j,premier_data,df):
   home_name=df.HomeTeam[i]
   dummy=[]
   counter=0
   for z  in range(i+1,3420):
     if(counter<10):
      if(premier_data.HomeTeam[z]==home_name):
        dummy.append(premier_data.iloc[:, j][z])
        counter=counter+1
     else:
       break
   if(len(dummy)>0):
    valor=statistics.mean(map(float,dummy))
   else:
    valor=-33
   return valor
```

```python
def away_mean(i,j,premier_data,df):
   away_name=df.AwayTeam[i]
   dummy=[]
   counter=0
   for z  in range(i+1,3420):
     if(counter<10):
      if(premier_data.AwayTeam[z]==away_name):
        dummy.append(premier_data.iloc[:, j][z])
        counter=counter+1
     else:
       break
   if(len(dummy)>0):
    valor=statistics.mean(map(float,dummy))
   else:
    valor=-33
   return valor
```

```python
for i in df.index:
  df.loc[i,'FTHG']= home_mean(i,3,premier_data,df)
  df.loc[i,'FTAG']= home_mean(i,4,premier_data,df)
  df.loc[i,'HTHG']= home_mean(i,6,premier_data,df)
  df.loc[i,'HTAG']= home_mean(i,7,premier_data,df)
  df.loc[i,'HS']= home_mean(i,9,premier_data,df)
  df.loc[i,'AS']= home_mean(i,10,premier_data,df)
  df.loc[i,'HST']= home_mean(i,11,premier_data,df)
  df.loc[i,'AST']= home_mean(i,12,premier_data,df)
  df.loc[i,'HF']= home_mean(i,13,premier_data,df)
  df.loc[i,'AF']= home_mean(i,14,premier_data,df)
  df.loc[i,'HC']= home_mean(i,15,premier_data,df)
  df.loc[i,'AC']= home_mean(i,16,premier_data,df)
  df.loc[i,'HY']= home_mean(i,17,premier_data,df)
  df.loc[i,'AY']= home_mean(i,18,premier_data,df)
  df.loc[i,'HR']= home_mean(i,19,premier_data,df)
  df.loc[i,'AR']= home_mean(i,20,premier_data,df)
  df.loc[i,'home_points']= home_mean(i,24,premier_data,df)
  df.loc[i,'away_points']= home_mean(i,25,premier_data,df)
  
```
